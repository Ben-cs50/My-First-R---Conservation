Importing the required libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the pre and post survey data
pre_data = pd.read_csv('Wild Tomorrow Survey Responses - Pre-Survey.csv')
post_data = pd.read_csv('Wild Tomorrow Survey Responses - Pre-Survey.csv')
```

```python
# Displaying duplicate entries
pre_data_duplicate = pre_data[pre_data.duplicated()]
pre_data_duplicate
```

```python
# Displaying duplicate entries
post_data_duplicate = post_data[post_data.duplicated()]
post_data_duplicate
```

This shows that we do not have duplicated rows

```python
# Cheking the total missing values
pre_data_missing = pre_data.isna().sum()
pre_data_missing
```

```python
# Cheking the total missing values
post_data_missing = post_data.isna().sum()
post_data_missing
```

```python
# Handling missing values by droping
pre_data = pre_data.dropna()
post_data = post_data.dropna()
```

Checking for missing values again

```python
# Checking missing values for pre-survey
pre_data_missing_after = pre_data.isna().sum()
pre_data_missing_after
```

```python
# Checking missing values for post-survey
post_data_missing_after = post_data.isna().sum()
post_data_missing_after
```

This shows that we have successfuly droped all the missing values
