Frequency Comparison: Compare the frequencies of responses (e.g., Yes/No answers) between pre and post surveys for specific questions related to conservation knowledge, attitudes, and behaviors.

Change Analysis: Calculate the change in responses for each individual between pre and post surveys. This can involve subtracting pre-survey responses from post-survey responses to identify changes in knowledge, attitudes, or behaviors.

Qualitative Analysis: Analyze open-ended responses or qualitative feedback provided in both pre and post surveys to identify common themes or trends.

```python
from scipy.stats import chi2_contingency

# Define a function to conduct chi-square test and visualize distributions for common fields
def compare_and_visualize(selected_pre_data, selected_post_data, column_name):
    if column_name in selected_pre_data.columns and column_name in selected_post_data.columns:
        # Conduct chi-square test
        contingency_table = pd.crosstab(selected_pre_data[column_name], selected_post_data[column_name])
        chi2, p, dof, expected = chi2_contingency(contingency_table)
        print(f"Chi-square statistic for '{column_name}': {chi2}")
        print(f"P-value for '{column_name}': {p}")

        # Visualize distributions
        pre_counts = selected_pre_data[column_name].value_counts()
        post_counts = selected_post_data[column_name].value_counts()

        plt.figure(figsize=(10, 5))
        plt.subplot(1, 2, 1)
        plt.bar(pre_counts.index, pre_counts.values, color='#117679')  # Pre-survey color
        plt.title(f'Pre-Survey {column_name}')
        plt.xlabel(column_name)
        plt.ylabel('Count')

        plt.subplot(1, 2, 2)
        plt.bar(post_counts.index, post_counts.values, color='#3F592E')  # Post-survey color
        plt.title(f'Post-Survey {column_name}')
        plt.xlabel(column_name)
        plt.ylabel('Count')

        plt.tight_layout()
        plt.show()
    else:
        print(f"'{column_name}' is not present in both pre-survey and post-survey data.")

# Conduct hypothesis testing and visualize distributions for common fields
common_fields = ['Gender', 'What is an ecosystem?', 'Is it good to kill insects?', 'Learner Age', 'Is conservation important for people?']
for field in common_fields:
    compare_and_visualize(selected_pre_comparable, selected_post_comparable, field)
```

# Step 8: Interpretaton and Reporting

To interpret the analysis findings and summarize the impact of lessons based on the analysis:

1. **Gender Distribution (Gender):**
   - The chi-square test results indicate that there is no significant difference in gender distribution between the pre and post surveys (Chi-square statistic = 0.1875, p-value = 0.665). This suggests that the lessons did not have a notable effect on the gender distribution of participants.

2. **Understanding of Ecosystems (What is an ecosystem?):**
   - The chi-square test results show no significant difference in responses related to understanding ecosystems between the pre and post surveys (Chi-square statistic = 0.0, p-value = 1.0). This implies that the lessons may not have significantly impacted participants' understanding of ecosystems.

3. **Attitude Towards Killing Insects (Is it good to kill insects?):**
   - The chi-square test results indicate no significant change in attitudes towards killing insects before and after the lessons (Chi-square statistic = 0.0, p-value = 1.0). This suggests that the lessons may not have influenced participants' attitudes towards insect conservation.

4. **Age Distribution (Learner Age):**
   - Although the chi-square test statistic for learner age is 6.0, indicating some discrepancy, the p-value of 0.199 suggests that the difference in age distribution between pre and post surveys is not statistically significant. This implies that the lessons likely did not target specific age groups differently.

5. **Awareness of Conservation Importance (Is conservation important for people?):**
   - Similar to other variables, the chi-square test results show no significant change in awareness of conservation importance between the pre and post surveys (Chi-square statistic = 0.0, p-value = 1.0). This indicates that the lessons may not have significantly impacted participants' awareness of conservation importance.

**Overall Summary:**
- The analysis suggests that the lessons may not have produced significant changes in the measured variables between the pre and post surveys. It is possible that additional factors beyond the scope of this analysis, such as lesson delivery, duration, or participant engagement, may have influenced the effectiveness of the lessons.
- Further evaluation, possibly through qualitative feedback or extended longitudinal studies, may provide deeper insights into the impact of conservation lessons on participants' knowledge, attitudes, and behaviors.
